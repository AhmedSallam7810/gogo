This Function is to validate if sequence is DNA or RNA or not

```python
def DNAValidation(dnaseq, seqtype):
    for n in dnaseq:
        if n not in seqtype:
            return False
    return True
```

This function is to count Neoclutides

```python
def countingnuc(seq):
    ns = {'A':0,
          'T':0,
          'C':0,
          'G':0
          }
    for n in seq:
        ns[n] = ns[n] +1
    return ns
```

DNA to RNA

```python
def DNAtoRNA(seq):
    return seq.replace('T', 'U')
```

Translating RNA to Protein

```python
#DNA--->RNA--->Amino Acid--->proteins
def Translating(seq):
    amino = ''
    for i in range(0,len(seq),3):
        if codons[seq[i:i+3]] != 'STOP':
            amino = amino +"-"+codons[seq[i:i+3]]
    return amino
```

Point Mutation

```python
def countMutation(s1,s2):
    m = 0
    for i in range(0,len(s2)):
        if s1[i] == s2[i]:
            continue
        else:
            m = m + 1
    return m
```

DNA complement

```python
def comp(dnaseq):
    c = ''
    for i in dnaseq:
        if i =='A':
            c = c + 'T'
        if i =='T':
            c = c + 'A'
        if i =='C':
            c = c + 'G'
        if i =='G':
            c = c + 'C'
    return c
```

DNA Reverse

```python
def reverse(dnaseq):
    c = ''
    d = {
        'A':'T',
        'T':'A',
        'C':'G',
        'G':'C'
    }
    for n in dnaseq:
        c = c + d[n]
    
    return c[::-1]
```
